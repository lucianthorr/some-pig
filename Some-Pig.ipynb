```python
from random import randint
from statistics import mean
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class Pig_Game:
    def __init__(self,player):
        self.player = player

    def roll_die(self):
        return randint(1,6)

    def pig_turn(self):
        turn_score = 0
        decision = "Play"
        while decision != "Hold":
            roll = self.roll_die()
            if roll == 1:
                self.player.turn_over()
                return 0
            else:
                turn_score += roll
                decision = self.player.decide(roll)
        self.player.turn_over()
        return turn_score

    def pig_game(self):
        score = 0
        for _ in range(7):
            score += self.pig_turn()
            
        self.player.end_game(score)
        return score
```

```python

class Player:
    """ This is the base player which always holds after the first roll. """

    def __init__(self):
        self.name = "Generic Player"
        self.score = 0
        self.rolls = 0
        self.turn_score = 0

    def decide(self, roll):
            self.score += roll
            return "Hold"

    def turn_over(self):
        pass

    def end_game(self,score):
        self.score = 0
        self.rolls = 0
        self.turn_score = 0

class Timid_Player(Player):

    def __init__(self):
        super().__init__()
        self.name = "Timid_Player"

    def decide(self, roll):
        self.score += roll
        self.rolls += 1
        if self.rolls > 2:
            return "Hold"

    def turn_over(self):
        self.rolls = 0


class Gamblers_Fallacy_Player(Player):
    def __init__(self):
        super().__init__()
        self.name = "Gamblers_Fallacy_Player"

    def decide(self, roll):
        self.score += roll
        self.rolls += 1
        if self.rolls > 4:
            return "Hold"

    def turn_over(self):
        self.rolls = 0


class Unpredictable_Player(Player):
    def __init__(self):
        super().__init__()
        self.name = "Unpredictable_Player"

    def decide(self, roll):
        self.score += roll
        self.rolls += 1
        if self.rolls > (randint(2,6)):
            return "Hold"

    def turn_over(self):
        self.rolls = 0


class Greedy_Player(Player):
    def __init__(self):
        super().__init__()
        self.name = "Greedy_Player"

    def decide(self, roll):
        self.score += roll
        self.turn_score += roll
        if self.turn_score > 20:
            self.turn_score = 0
            return "Hold"

    def turn_over(self):
        self.turn_score = 0
        
class Smart_Player(Player):
    def __init__(self):
        super().__init__()
        self.name = "Smart_Player"
        self.max_turn_score = randint(1,25)
        self.games_played = 0
        self.tried_scores = {}
        
    
    def find_best_score(self):
        best_average = 0
        best_turn = 0
        for turn_score, end_scores in self.tried_scores.items():
            if mean(end_scores) > best_average:
                best_average = mean(end_scores)
                
                best_turn = turn_score
        return best_turn

    def decide(self, roll):
        self.score += roll
        self.turn_score += roll
        if self.games_played == 200:
            self.games_played+=1
            self.max_turn_score = self.find_best_score()
        if self.turn_score > self.max_turn_score:
            #self.turn_score = 0
            return "Hold"

    def turn_over(self):
        self.turn_score = 0
        
    def end_game(self,score):
        self.score = score
        try:
            self.tried_scores[self.max_turn_score].append(score)
        except:
            self.tried_scores[self.max_turn_score] = [score]
        if(self.games_played < 200):
            self.max_turn_score = randint(1,25)
        self.games_played += 1
        self.score = 0
        self.rolls = 0
        self.turn_score = 0
        

```

```python
smart_player = Smart_Player()
game = Pig_Game(smart_player)
score = game.pig_game()
```

```python
class Simulator:

    def __init__(self):
        pass

    def trial(self,player):
        game = Pig_Game(player)
        score = game.pig_game()
        return score

    def run_trials(self,player,num_of_trials):
        scores = []
        for _ in range(num_of_trials):
            scores.append(self.trial(player))
        return scores
```

### Plot Scores of Each Player over 1000 Games

```python
simulator = Simulator()
generic_scores = simulator.run_trials(Player(),1000)
timid_scores = simulator.run_trials(Timid_Player(),1000)
gamblers_scores = simulator.run_trials(Gamblers_Fallacy_Player(),1000)
unpredictable_scores = simulator.run_trials(Unpredictable_Player(),1000)
greedy_scores = simulator.run_trials(Greedy_Player(),1000)
smart_scores =  simulator.run_trials(Smart_Player(),1000)
```

```python
plt.figure(figsize=(10,10))
plt.plot(generic_scores,label="Generic")
plt.plot(timid_scores,label="Timid")
plt.plot(gamblers_scores,label="Gambler")
plt.plot(unpredictable_scores,label="Unpredictable")
plt.plot(greedy_scores,label="Greedy")
plt.plot(smart_scores,label = "Smart")
plt.legend()
plt.show()
```

### Did Smart Player improve over time? (Compare with Greedy Player, which seemed to be best)

```python
generic_scores = simulator.run_trials(Player(),10000)
smart_scores = simulator.run_trials(Smart_Player(),10000)
greedy_scores = simulator.run_trials(Greedy_Player(),10000)
gambler_scores = simulator.run_trials(Gamblers_Fallacy_Player(),10000)
smart_averages = []
greedy_averages = []
generic_averages = []
gambler_averages = []
plt.ylabel("Average Score")
plt.xlabel("Games Played")
for index, scores in enumerate(smart_scores):
    if index % 50 == 0 and index != 0:
        smart_averages.append(mean(smart_scores[0:index]))
for index, scores in enumerate(greedy_scores):
    if index % 50 == 0 and index != 0:
        greedy_averages.append(mean(greedy_scores[0:index]))
for index, scores in enumerate(generic_scores):
    if index % 50 == 0 and index != 0:
        generic_averages.append(mean(generic_scores[0:index]))
for index, scores in enumerate(gambler_scores):
    if index % 50 == 0 and index != 0:
        gambler_averages.append(mean(gambler_scores[0:index]))
        
plt.plot([x for x in range(1,10000) if x % 50 == 0],smart_averages,label="Smart")
plt.plot([x for x in range(1,10000) if x % 50 == 0],greedy_averages,label="Greedy")
plt.plot([x for x in range(1,10000) if x % 50 == 0],generic_averages,label="Generic")
plt.plot([x for x in range(1,10000) if x % 50 == 0],gambler_averages,label="Gambler")
plt.legend()
plt.show()
```
